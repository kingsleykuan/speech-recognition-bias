```python
Crema = "./Data/AudioWAV/"
```

```python
crema_directory_list = os.listdir(Crema)

file_emotion = []
file_path = []
actor_id = []

for file in crema_directory_list:
    # storing file paths
    file_path.append(Crema + file)
    # storing file emotions
    part=file.split('_')
    file_emotion.append(part[2])
    actor_id.append(int(part[0][-4:]))

#Construct Dataframe
emotion_df = pd.DataFrame(file_emotion, columns=['Emotion'])
path_df = pd.DataFrame(file_path, columns=['Path'])
actor_df = pd.DataFrame(actor_id, columns = ["ActorID"])
temp_df = pd.concat([actor_df,emotion_df, path_df], axis=1)

crema_df = temp_df.merge(demo_df, on = "ActorID")
crema_df.head()
```
