# Prepare Env

```python
import os
import sys
# Make sure you are running in the project folder
ROOT_FOLDER = os.getcwd() + "/web/"
DATA_SET_FOLDER = ROOT_FOLDER + "/test/"
RESULTS_FOLDER = ROOT_FOLDER + '/results/'
os.chdir(ROOT_FOLDER)
sys.path.append(os.path.abspath(ROOT_FOLDER))
print("Working dir: " + ROOT_FOLDER)

wav_seq = []
for root, dirs, files in os.walk(DATA_SET_FOLDER):
    for filename in files:
        wav_seq.append(filename)
print("Reading wav file number:", len(wav_seq))

import audio_handler as ah
import commercial_api as ca 
```

# Run

```python
check_point = 0
for i in range(check_point, len(wav_seq)):
    wav = ah.readWavFromPath(DATA_SET_FOLDER + wav_seq[i])
    index = wav_seq[i].split('.')[0] # Data index
    audio_format = 'wav' # ogg also supported
    print("Current index : ",i, " Name: ", index)
    result = ca.emotion_detect(wav, index, audio_format)
    with open(RESULTS_FOLDER + index + '.json', 'w') as f:
        f.write(result)
        f.close()
    print('\n')
```

```python
## Replace empath
import json
p = 640
for i in range(p, len(wav_seq)):
    wav = ah.readWavFromPath(DATA_SET_FOLDER + wav_seq[i])
    index = wav_seq[i].split('.')[0]
    audio_format = 'wav'
    print("Current index : ",i, " Name: ", index)
    result = ca.emotion_detect(wav, index, audio_format)
    print(result)
    j = None
    with open(RESULTS_FOLDER + index + '.json', 'r') as f:
        j = json.loads(f.read())
        print(j)
        f.close()
        
    with open(RESULTS_FOLDER + index + '.json', 'w') as f:
        j['empath'] = json.loads(result)['empath']
        f.write(json.dumps(j))
        f.close()
    print('\n')
```

```python
# Do check
import json
count = 0
for root, dirs, files in os.walk(RESULTS_FOLDER):
    for filename in files:
        with open(RESULTS_FOLDER + filename, 'r') as f:
            try:
                content = f.read()
                j = json.loads(content)
                if len(content) == 0:
                    print(filename)
                if 'Happy' not in j['vokaturi']:
                    print("bad: ", filename)
                    count += 1
            except:
                pass
print("Checking done, bad: ", count)
```
